### Regularization

Regularization is a technique to deal with overfitting. 

In terms of logistic regression it is given by 

L2 Regularization,

$$ L2 = ||W||^2 =  \sum_{j=1}^{n_x} w_j^2  = w^T . w  $$ 

and Loss function would be given as,

$$ J(w,b) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{m} ||w||^2_2 $$

where $ \lambda $ is a regularization parameter

L1 regularization,

$$ L1 =  \sum_{j=1}^{n_x} |w_j|  $$

and loss function would be given as,

$$ J(w,b) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{2m} ||w||_1 $$

In this case, it is observed that the w data will be sparsed meaning there would be many zero values. 



In terms of Neural Network,

L2 Regularization is given as,

$$ L2 = ||W||^2 =  \sum_{i=1}^{n^{[l]}} \sum_{j=1}^{n^{[l]-1}} w_{ij}^{[l]2} $$ 

and loss functions can be given as,

$$ J(w,b,...,w^{[l]},b^{[l]}) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{2m} \sum_{i=1}^{i=l}||w^l||^2_F $$

It is called Frobenius norm.

So with Regularization the idea is to narrow down the values of certain weights or maybe bring them close to zero such that they have very negligible impact on the neural network. By increasing the value of $\lambda $ we can zero certain weights and hence regularize the neural network. This will bring down the problem of overfitting. However, in reality it is not exactly zeroing down the weights but infact making the value so small close to zero such that they have very minimal impact on the neural network and it behaves more like simple neural network. 

Another intuition about making the weight values bring down close to zero by cracking up the value of $\lambda $ can be taken from the activation function of tanh. 

If the weights are small, the value of Z is small and hence the activation function of tanh nearly follows the linear regime as the value of z stays around the linear area. Hence the network moresoever performs like a linear neural network in some layers. Fundamentally, the effect of linear boundary is that is does not fit the data perfectly and it misclassifies some points, thus reducing the overfitting. However, the whole idea again is to tweak the value such that the data does not overfit and optimize the network to have low variance and low bias. 




### Dropout Regularization

Dropout regularization is only used at trainig time and not at test time. The general idea of dropout is that we are eliminating few neurons in a layer thus making the neural network much simpler. 

Dropout regularization works on the principle of keeping the probablities of having certain nodes and eliminating others. For example a probability of 0.8 would mean to keep 80% of the nodes in a certain layer and dropping 20% of them. This probability can be kept different for different layers.




### Data Augmentation

Sometimes it occurs that in order to reduce the overfitting we need to add more training examples, however if it is difficult to add independent training examples, the images can be flipped horizontally and that can be added as a new example to increase the size of the dataset.

For optical augmentation, we can also rotate, distort a number for example to add to the dataset.  



#### Early Stopping

Early stopping is another regularization technique which specifically focuses on stopping the iterations earlier before the test set error starts to climb again after a point of deep when even though the cost function or training error is continuously decreasing. This is another way to cope up with overfitting when you can not get access to more data. It is a good idea to stop at the iterations where the dev set error is minimum.
