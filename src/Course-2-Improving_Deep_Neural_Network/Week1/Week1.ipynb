### Regularization

Regularization is a technique to deal with high variance classifier. 

In terms of logistic regression it is given by 

L2 Regularization,

$$ L2 = ||W||^2 =  \sum_{j=1}^{n_x} w_j^2  = w^T . w  $$ 

and Loss function would be given as,

$$ J(w,b) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{m} ||w||^2_2 $$

where $ \lambda $ is a regularization parameter

L1 regularization,

$$ L1 =  \sum_{j=1}^{n_x} |w_j|  $$

and loss function would be given as,

$$ J(w,b) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{2m} ||w||_1 $$

In this case, it is observed that the w data will be sparsed meaning there would be many zero values. 



In terms of Neural Network,

L2 Regularization is given as,

$$ L2 = ||W||^2 =  \sum_{i=1}^{n^{[l]}} \sum_{j=1}^{n^{[l]-1}} w_{ij}^{[l]2} $$ 

and loss functions can be given as,

$$ J(w,b,...,w^{[l]},b^{[l]}) = \frac{1}{m} \sum_{i=1}^{m} L(\hat{y}^i,y^i) +  \frac{\lambda}{2m} \sum_{i=1}^{i=l}||w^l||^2_F $$

It is called Frobenius norm.

So with Regularization the idea is to narrow down the values of certain weights or maybe bring them close to zero such that they have very negligible impact on the neural network. By increasing the value of $\lambda $ we can zero certain weights and hence regularize the neural network. This will bring down the problem of overfitting.
