## Activation functions:

When building a neural network we can also choose different activation functions for the hidden layer as well output layers. So far we have only used sigmoid functions for the examples before and for programming assingments. 

#### Sigmoid and tanh functions comparision







tanh activation function can be given by 

$$ g(z) = tanh(z) = \frac{e^{z} - e^{-z}}{e^{z} + e^{-z}} $$

where g will be used as a notation for global activation function. 


#### ReLu Activation function

Rectified Linear Unit is the full form of Relu function. In using both sigmoid and tanh functions the gradient descent slows down when the value of z approaches very low. This inturn increases the computaion time. An alternative function to this is ReLu function, where any negative value of zero gives zero slope while any value bigger than z gives the slope of z itself. 

$$ a = max(0,z)$$



### Derivatives of Activation Function

Derivative of sigmoid function:

the sigmoid function is given as  : $ \sigma = \frac {1} {1+e^-z} $  and derivative of this function comes out to be $$g'(z) = a = g(z)(1-g(z)) $$


Derivative of tanh function:
 
the tanh function is given as : $ tan(z) = \frac {e^z -e ^{-z}} {e^z + e^{-z}} $ and derivative of this function comes out to be

$$ g'(z) = 1 - (tanh(z))^2 $$

and now we have $$ a = g(z) =  tanh(z) $$ then $$g'(z) = 1- a^2 $$


Derivative of ReLu:

The function is given by  $g(z) = max(0,z) $, hence

$$ g'(z) = 0 $$ if $$  z<0 $$
and $$g'(z) = 1 $$ if $$ z>0 $$
