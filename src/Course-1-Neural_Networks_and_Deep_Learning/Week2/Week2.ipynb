#### Binary Classification

Notations:

y belongs to a set of 0 and 1. 


#### Logistic Regression

In logistic regression let's say we have x as input and y-hat as prediction then y-hat would be the probability of y being 1 given the input x. 

$$ \hat{y} = P(y=1 | x) $$

Now, the logistic regression will have parameters W and b, b being the real number and W being a vector. 

The output function would be  $$ \hat{y}= W^T + b $$


However, in binary classification we would need $\hat{y}$ to be between 0 and 1. So we could not use the linear function as it may give negative outputs as well. Hence we multiply the term by $ \sigma $. 

$$ \hat{y} =  \sigma (W^T + b) $$

Cost function in Logistic Regression:

Loss function is regression is normally given as the sqaured error as,

Loss function $ = 1/2 (\hat{y}-y)^2 $

However this function does not provide the convex shape for gradient descent we aim for and we cannot find the best optimization at minimum error. This is because we are applying this loss function to a binary classification problem. 

For logistic regression , the loss function can be given as,

$$ L_f = - y log\hat{y} + (1-y) log (1-\hat{y}) $$

and generalized form for multiple training examples will be called Cost function can be given as average of the sum of all loss functions of training examples. 

$$ C_f = - \frac{1}{m} \sum_{i=1}^{m} y^i log\hat{y}^i + (1-y^i) log (1-\hat{y}^i)$$

where m is the total number of training examples.


#### Gradient Descent

Gradient descent is an important term that is used in determining loss function step by step. 

If we have one dimensional problem , then the Loss function is a function of Weights and bias of one dimension. 

The idea here is to the initialize W and b with some values, could be zero or it could be any other value. Then measure the cost function for that step. The initialized values will not give the minimum cost function. The idea is to keep updating the weights and bias till the minimum point is reached on the convex curve where the error is minimum. This process is called Gradient descenet. 

The amount by with the weights needs to be updated is given by learning rate multiplied partial differential of cost function. 

$$ \omega - \alpha  \frac{\partial C_{(W,b)}}{\partial \omega}$$
